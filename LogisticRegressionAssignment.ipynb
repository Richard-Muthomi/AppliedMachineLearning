**Data Cleaning**

```python
imr['Value'].mean()
```

```python
def impute_value(cols):
  Value=cols[0]
  Year=cols[1]
  if pd.isnull(Value):
    return 69.3324377137274  # Impute with the mean for missing values which is Value.mean()
  else:
    return Value
```

```python
imr['Value']= imr[['Value','Year']].apply(impute_value,axis=1)
```

```python
series=pd.get_dummies(imr['Series'],drop_first=True)
```

```python
imr=pd.concat([imr,series,source],axis=1)
```

```python
#Build Logistic Regression Model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(imr.drop('Other of concern to UNHCR (number)',axis=1),imr['Other of concern to UNHCR (number)'],
                                                 test_size=0.20,random_state=101)
```

```python
#Training and Predicting
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel= LogisticRegression()
```

```python
logmodel
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
predictions
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
print(classification_report(y_test,predictions))
```

```python
print(sns.heatmap(confusion_matrix(predictions,y_test),annot=True))
```
